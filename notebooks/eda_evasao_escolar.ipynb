```python
# Análise Exploratória de Dados - Evasão Escolar

import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados
df = pd.read_csv("evasao_escolar.csv")

# Visualizar as primeiras linhas
df.head()
```

```python
# Informações gerais do dataset
df.info()
```

```python
# Estatísticas descritivas
df.describe()
```

```python
# Verificar valores ausentes
df.isna().sum()
```

```python
# Taxa média de evasão por região
evasao_regiao = df.groupby("regiao")["taxa_evasao"].mean()
evasao_regiao
```

```python
# Gráfico de evasão por região
evasao_regiao.plot(kind="bar", title="Taxa Média de Evasão por Região")
plt.ylabel("Taxa de Evasão")
plt.show()
```

```python
# Evolução da evasão ao longo do tempo
evasao_tempo = df.groupby("ano")["taxa_evasao"].mean()
evasao_tempo.plot(title="Evolução da Evasão Escolar ao Longo do Tempo")
plt.ylabel("Taxa de Evasão")
plt.show()
```

```python
# Salvar dados tratados
df.to_csv("evasao_tratada.csv", index=False)
```
